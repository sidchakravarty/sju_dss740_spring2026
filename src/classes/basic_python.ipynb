```python
def calculate_squareroot(x: float) -> float:
    """Calculate the square root of a number."""
    return np.sqrt(x)
```

```python
original_value = 16.0
square_root = calculate_squareroot(original_value)
print(f"The square root of {original_value} is {square_root}")
```

```python
original_value = 239.45
square_root = calculate_squareroot(original_value)
print(f"The square root of {original_value} is {square_root}")
```
